# Append Citi Bike datasets into one document for use in Tableau

```python
# Load Module

import pandas as pd
```

```python
# File Paths

file_one = "Resources/2014-01-CitiBikeTripData.csv"
file_two = "Resources/2014-02-CitiBikeTripData.csv"
file_three = "Resources/2014-03-CitiBikeTripData.csv"
file_four = "Resources/2014-07-CitiBikeTripData.csv"
file_five = "Resources/2014-08-CitiBikeTripData.csv"
file_six = "Resources/2014-09-CitiBikeTripData.csv"
```

```python
# Read Files

file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")
file_two_df = pd.read_csv(file_two, encoding="ISO-8859-1")
file_three_df = pd.read_csv(file_three, encoding="ISO-8859-1")
file_four_df = pd.read_csv(file_four, encoding="ISO-8859-1")
file_five_df = pd.read_csv(file_five, encoding="ISO-8859-1")
file_six_df = pd.read_csv(file_six, encoding="ISO-8859-1")
```

```python
# Append Files

append = file_one_df.append([file_two_df, file_three_df, file_four_df, file_five_df, file_six_df])

append.head()
```

```python
# Export File

# Export file as a CSV, without the Pandas index, but with the header
append.to_csv("Output/WinterSummer2014CitiBikeData.csv", index=False, header=True)
```
